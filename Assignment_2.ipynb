```python
import seaborn as sns
```

##**1)SCATTERPLOT**

```python
sns.scatterplot(x="alcohol",y="speeding",data=dataset)
```

##**Inference of Scatterplot**
#Taking alcohol on x-axis and speeding on y-axis
#A sctterplot is obtained
#The above scatterplot depicts the relationship between two variables in the car crashes dataset: "alcohol" and "speeding." Each point on the scatterplot represents a data point from the dataset (most likely a car crash incident).

##**2)LINEPLOT**

```python
sns.lineplot(x="alcohol",y="total",data=dataset,ci=None)
```

##**Inference of Lineplot**
#Taking alcohol on x-axis and total on y-axis
#A lineplot is obtained
#The line plot depicts the distribution of "total" car crashes in cases involving and not involving alcohol. You can see if there is a significant difference in the average or total number of car crashes between cases involving and not involving alcohol.


##**3)DISTRIBUTION PLOT**

```python
sns.distplot(dataset["speeding"])
```

##**Inference of Distribution plot**
#A distribution plot for "speeding" attribute is obtained
#We can draw conclusions about the prevalence of speeding in car accidents based on the dataset, such as whether it follows a normal distribution, whether there are prominent peaks or clusters of speeding incidents, and whether there are any outliers or unusual patterns.  


##**4)HISTOGRAM**

```python
sns.histplot(data=dataset, x='alcohol', y='speeding')
```

##**Inference of Histogram**
#Taking "alcohol" attribute on x-axis and "speeding" attribute on y-axis
#A histogram is plotted
#Gaps or empty regions in the histogram, which could indicate missing or incomplete data.

##**5)RELATIONPLOT**

```python
#using categorical attribute as hue
sns.relplot(x="alcohol",y="speeding",data=dataset,hue="abbrev")
```

##**Inference of Relation plot**
#Taking alcohol on x-axis and speeding on y-axis
#A relation plot is obtained
#The relational plot (relplot) generates a scatter plot to depict the relationship between "alcohol" and "speeding" involvement in car accidents, with points coloured by the "abbrev" variable (likely representing different states or regions).Plot enables you to identify potential trends, differences, or outliers in the dataset by allowing you to understand the interplay between alcohol and speeding in car crashes across different states or regions.

##**6)BARPLOT**

```python
sns.barplot(data=dataset,x="alcohol",y="speeding",ci=None)
```

##**Inference of Barplot**
#A barplot is obtained with "alcohol" on x-axis and "speeding" on y-axis
#This bar plot compares the average levels of speeding involvement in car crashes with and without alcohol involvement, providing insight into the potential relationship between these two factors.

##**7)COUNTPLOT**

```python
sns.countplot(x="speeding",data=dataset)
```

##**Inference of a countplot**
#A countplot is obtained for "speeding" attribute
#This graph depicts the distribution of car accidents involving and not involving speeding. It enables you to estimate the frequency of speeding-related crashes.

##**8)JOINTPLOT**

```python
sns.jointplot(x="alcohol",y="speeding",data=dataset)
```

##**Inference of a Jointplot
#A jointplot is obtained by taking "alcohol" on x-axis and "speeding" on y-axis
#The distribution of both variables is depicted in the plot, indicating the frequency of alcohol and speeding involvement. The scattered points and histograms aid in the identification of any patterns or correlations between the two variables, providing insight into their potential association in car accidents.


##**9)BOXPLOT**

```python
sns.boxplot(x="alcohol",y="speeding",data=dataset)
```

##**Inference of Boxplot**
#A boxplot is obtained with "alcohol" on x-axis and "speeding" on y-axis
#This box plot compares the distribution, central tendency, and spread of speeding involvement in car accidents with and without alcohol involvement, assisting you in understanding potential differences or patterns between these two factors.

##**10)CORRELATION**

```python
corr=dataset.corr()
corr
```

#**Inference of Correlation**
#The correlation matrix provides valuable preliminary insight into the relationships between numeric variables in the dataset, assisting you in identifying potential patterns or dependencies that warrant further investigation.

##**11)HEATMAP**

```python
sns.heatmap(corr,annot=True,cmap="YlGnBu")
```

##**Inference of Heatmap**
#The heatmap depicts the correlation structure within the dataset in a clear visual manner, assisting in the identification of relationships between variables and informing data pre-processing and modelling decisions.
